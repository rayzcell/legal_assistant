```python
#kb sample
kb = [
    "Related cases for breach of contract in corporate law",
    "Breach of contract dispute resolution for service agreements",
    "Employment discrimination cases involving gender bias",
    "Intellectual property dispute cases for patent infringement",
    "How to search for cases involving wrongful termination",
    "Search phrases for cases related to breach of contract in construction law",
    "Precedents for intellectual property disputes involving software patents",
    "Strategies for finding past rulings on employment discrimination",
    "Corporate law: cases involving breach of non-compete agreements",
    "Legal strategies for handling intellectual property theft"
]


```

```python
import faiss
from sentence_transformers import SentenceTransformer

# for embedding Load pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# KB embeddings for kb
kb_embeddings = model.encode(kb)

# Createng FAISS index for vrctorized KB
index = faiss.IndexFlatL2(kb_embeddings.shape[1])
index.add(kb_embeddings)
```

```python
#user query
new_query = "I need help with breach of contract in corporate law"
query_embedding = model.encode([new_query])

# Searchng FAISS index for the most relevant logs
k = 3
distances, indices = index.search(query_embedding, k)

# retriving  top search logs
retrieved_logs = [kb[idx] for idx in indices[0]]
print("Retrieved Logs:", retrieved_logs)
```

```python
import json

def preprocess_data(data):
    categories = {}
    docs = []

    # Process categories
    for category in data['categories']:
        category_name = category[0]
        category_items = category[1]
        categories[category_name] = [{'value': item['value'], 'formInput': item['formInput']} for item in category_items]

    # Process documents
    for doc in data['docs']:
        doc_info = {
            'id': doc['tid'],
            'title': doc['title'],
            'publish_date': doc['publishdate'],
            'author': doc.get('author', 'Unknown'),
            'doc_source': doc['docsource'],
            'num_cited_by': doc['numcitedby'],
            'headline': doc['headline'],
            'fragment': doc['fragment']
        }
        docs.append(doc_info)

    return {
        'categories': categories,
        'documents': docs,
        'total_documents': data['found']
    }


api_token = "6ac69313684dfe82332e7949d640bea1dc931272"

headers = {
    'Authorization': f'Token {api_token}'
}

query = "freedom of speech"
url = f"https://api.indiankanoon.org/search/?formInput={query}&pagenum=0"

# Send a GET request
response = requests.post(url, headers=headers)

preprocess_response = preprocess_data(response.json())
```

```python
import gradio as gr
api_key = "6ac69313684dfe82332e7949d640bea1dc931272"
# Function to summarize legal data using GPT-2
def summarize_legal_case(query):
    case_text = fetch_case_data(query)

    if not case_text:
        return "No case found for the given query."

    preprocessed_text = preprocess_case_text(case_text)

    # Tokenize the input text for GPT-2
    inputs = tokenizer.encode(preprocessed_text, return_tensors="pt")
    outputs = model.generate(inputs, max_length=200, num_return_sequences=1, do_sample=True)

    # Decode and return the summary
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Create a Gradio interface for user input
interface = gr.Interface(fn=summarize_legal_case, inputs="text", outputs="text",
                         title="Legal Case Summarizer",
                         description="Enter a legal query, and the system will summarize relevant cases.")

interface.launch(debug=True)
```
